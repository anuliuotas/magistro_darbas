```python
hourly_data_df = pd.read_csv("Magistro duomenys/processed/model_data.csv", sep=',', decimal=',')
hourly_data_df['Time'] = pd.to_datetime(hourly_data_df['Time'])
hourly_data_df = hourly_data_df.set_index('Time')
hourly_data_df.index = pd.DatetimeIndex(hourly_data_df.index).to_period('M')
hourly_data_df = hourly_data_df.drop(['Unnamed: 0'], axis=1)
hourly_data_df = hourly_data_df.astype(float)
%store hourly_data_df
```

```python
import statistics

class DataAggregator:

    def mean(self, array):
        if len(array) == 0:
            return 0
        return sum(array) / len(array)

    def sum(self, array):
        return sum(array)

    def min(self, array):
        return min(array)

    def max(self, array):
        return max(array)

    def median(self, array):
        return statistics.median(array)

    def work_hours_mean(self, array):
        arr = array[7:16]
        return self.mean(arr)

    def night_hours_mean(self, array):
        arr = (array[0:6]).append(array[18:23])
        return self.mean(arr)

    def max_values_average(self, array):
        count = 5
        copy = array
        arr = []
        print(array)
        for i in range(0, count):
            mv = self.max(copy)
            arr.append(mv)

            copy = copy.drop(mv)
        return self.mean(arr)
```
