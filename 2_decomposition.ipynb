```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

import statsmodels.api as sm
```

```python
%store -r data_df
```

```python
series = data_df['Nord Pool Lietuva']
```

## Decomposition period = 365days

MetinÄ— dekompozicija

```python
res_1 = sm.tsa.seasonal_decompose(series,
                                period=365,
                                model='additive')
```

```python
fig, (ax0, ax1,ax2,ax3) = plt.subplots(4,1, figsize=(15,13))
ax0.plot(series)
ax1.plot(res_1.trend)
ax2.plot(res_1.seasonal)
ax3.plot(res_1.resid)

ax0.title.set_text('Observed')
ax1.title.set_text('Trend')
ax2.title.set_text('Seasonal')
ax3.title.set_text('Residual')
```

## Decomposition period = 365 days + 7days

MetinÄ—s dekompozicijos liekanos savaitinÄ— dekompozicija

```python
res_2 = sm.tsa.seasonal_decompose(res_1.resid.dropna(),
                                period=7,
                                model='additive')
```

```python
fig, (ax0, ax1,ax2,ax3) = plt.subplots(4,1, figsize=(15,13))
ax0.plot(series)
ax1.plot(res_2.trend)
ax2.plot(res_2.seasonal)
ax3.plot(res_2.resid)

ax0.title.set_text('Observed')
ax1.title.set_text('Trend')
ax2.title.set_text('Seasonal')
ax3.title.set_text('Residual')
```

## Decomposition period = 7 days

SavaitinÄ— dekompozicija

```python
res_3 = sm.tsa.seasonal_decompose(series,
                                period=7,
                                model='additive')
```

```python
fig, (ax0, ax1,ax2,ax3) = plt.subplots(4,1, figsize=(15,13))
ax0.plot(series)25
ax1.plot(res_3.trend)
ax2.plot(res_3.seasonal)
ax3.plot(res_3.resid)

ax0.title.set_text('Observed')
ax1.title.set_text('Trend')
ax2.title.set_text('Seasonal')
ax3.title.set_text('Residual')
```

## RezultatÅ³ suraÅ¡ymas

```python
data_df['decomposition_residuals_365d'] = res_1.resid
data_df['decomposition_residuals_365d_7d'] = res_2.resid
data_df['decomposition_residuals_7d'] = res_3.resid
```
